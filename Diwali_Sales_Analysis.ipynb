## Data Visualization

```python
gender_sales = df.groupby('Gender')['Amount'].sum()

gender_sales.plot(kind='bar')

plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales Amount")

plt.show()
```

```python
age_sales = df.groupby('Age Group')['Amount'].sum()

age_sales.plot(kind='bar')

plt.title("Total Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales")

plt.show()
```

```python
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False)

state_sales.head(10).plot(kind='bar')

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")

plt.show()
```

```python
occupation_sales = df.groupby('Occupation')['Amount'].sum()

occupation_sales.plot(kind='bar')

plt.title("Sales by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Total Sales")

plt.show()
```

```python
product_sales = df.groupby('Product_Category')['Amount'].sum()

product_sales.plot(kind='bar')

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")

plt.show()
```

## Insights from Diwali Sales Data Analysis

1. Female customers contributed more to the total sales compared to male customers.

2. The age group 26–35 made the highest number of purchases during the Diwali season.

3. States like Uttar Pradesh, Maharashtra, and Karnataka generated higher sales.

4. Working professionals and IT sector employees showed strong purchasing power.

5. Certain product categories such as Food, Clothing, and Electronics were among the most popular purchases during the festival.
